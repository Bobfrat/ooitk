```python
# Make it so notebook knows how to access ooitk from the notebook dir
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()))
```

```python
erddap_url = 'http://erddap-test.oceanobservatories.org:8080/erddap/tabledap/e1fd3967aa55475e8338fc36a6493f5c'
```

```python
from ooitk.session import ERDDAPSession

session = ERDDAPSession(erddap_url)
session.open()
```

```python
pres = session.variables['sci_water_pressure'][:]
temp = session.variables['sci_water_temp'][:]
cond = session.variables['sci_water_cond'][:]
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_autoscaley_on(False)
ax1.set_xlabel('Temperature (deg_C)', fontsize=14)
ax1.set_ylabel('Pressure (dbar)', fontsize=14)
ax1.set_ylim((-100,0))
ax1.scatter(temp,-pres, c='b', marker='.')

ax2.set_autoscaley_on(False)
ax2.set_xlabel('Conductivity (s m-1)', fontsize=14)
ax2.set_ylabel('Pressure (dbar)', fontsize=14)
ax2.set_ylim((-100,0))
ax2.scatter(cond, -pres,c='r', marker='D')

plt.show()
```
